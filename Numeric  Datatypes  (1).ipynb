# Numerical Datatype
 Numerical data refers to the data that is in the form of numbers
 There are three distinct numeric types
 * integers
 * float
 * complex numbers
 

```python
x,y,z=1234,0.96,5+8j
print(type(x), type(y), type(z))  # 'x' is integer, 'y' is float,'z' is complex number 
```

### Integers
An integer can be zero,postivr or a negative whole number 
 Ex:1234, 0, -55 

```python
x,y,z=1234,0,-55
print(type(x),type(y),type(z)) # here int is integer
```

* Integers are immutable

```python
x=5
print(id(x))
x=10
print(id(x)) # immutable (memory locations are different)
```

* Integers can be binary,octal and hexa decimal values

### Binary Case
A number consistiong of only the binary digits one(1) and zero(0) and prefixed with 0b is a binary number
* To represent an integer in binary form, store it directly as a literal,or use int() function, in which the base is set to '2'


```python
print(int('10')) #base '10'
print(int('0101', 2)) # binary case( base'2')
```

### Octal Numbers
A number system which has its base as eight(8) is called an Octal number 


```python
print(int('253',8)) # octal number (base '8')
```

### Hexadecimal Numbers
A Number system which has its base as '16'



```python
print(int('1AB',16))
```

### Float
A Floating point number, is a positive or negative whole number with decimal point

```python
float(5)
```

### Round
Round function is used for rounding off numbers upto specified number of digits after decimal point


### Complex Numbers
The complex data type in python consist of two values, the first one is the real part of the complex number and tthe second one is the imaginary part of the complex number  Ex:5+3j

```python
x=5+3j
print(x,type(x))
print(x.real, x.imag)
```

### Absolute value
The absolute valueof any number is always positive, it removes the negative sign of a number

```python
x=-10 #integer
print(abs(x))
```

```python
x=-0.97 #float
print(abs(x))
```

```python
x=-9-8j #complex number
print(abs(x))
```

### Conjugate complex number
The conjugate complex number is obtained by changing the sign of its imaginary part 

```python
x=5-3j
print(type(x))
```

### Floor and Ceil
### Floor 
Floor function rounds a number down to the nearest integer 
### Ceil
Ceil function rounds a number upwards to its nearest integer 

```python
import math
from math import ceil,floor
print(floor(5.345),ceil(5.345))
```

```python
int(5.345)+1 # without math getting ceil function
```
